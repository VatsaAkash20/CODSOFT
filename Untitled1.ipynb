```python
# TITANIC SURVIVAL PREDICTION PROJECT

# Step 1: Import necessary libraries
import pandas as pd
from sklearn.linear_model import LogisticRegression
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the dataset
data = pd.read_csv("Titanic-Dataset.csv")

# Step 3: Data cleaning
data = data.drop("Cabin", axis=1)  # Drop column with too many missing values
data["Age"].fillna(data["Age"].median(), inplace=True)
data["Embarked"].fillna(data["Embarked"].mode()[0], inplace=True)

# Step 4: Convert text data to numbers
data["Sex"] = data["Sex"].map({"male": 0, "female": 1})
data["Embarked"] = data["Embarked"].map({"S": 0, "C": 1, "Q": 2})

# Step 5: Select input features
features = ["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]
X = data[features]
y = data["Survived"]

# Step 6: Build and train the model
model = LogisticRegression(max_iter=200)
model.fit(X, y)

# Step 7: Predict survival for all 891 passengers
predictions = model.predict(X)

# Step 8: Add predictions to original data
data["Predicted_Survival"] = predictions

# Step 9: Print all passenger info with prediction
pd.set_option("display.max_rows", 891)
print(data[["PassengerId", "Name", "Sex", "Age", "Pclass", "Survived", "Predicted_Survival"]])

# Step 10: Basic analysis: Who survived more?
sns.barplot(x="Sex", y="Survived", data=data)
plt.title("Survival Rate by Gender (1 = female)")
plt.show()

sns.barplot(x="Pclass", y="Survived", data=data)
plt.title("Survival Rate by Class")
plt.show()
```
